```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
model, scaler, saved_cols = joblib.load("/content/diabetes_lr.pkl")
```

```python
df_test = pd.read_csv("/content/Train - Sheet1.csv")

cols =  ['Pregnancies', 'Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
df_test = df_test[cols]
```

```python
df_test.isnull().sum()
```

```python
df_test.fillna({'Glucose': df_test['Glucose'].mean()}, inplace=True)
df_test.fillna({'BloodPressure': df_test['BloodPressure'].mean()}, inplace=True)
df_test.fillna({'SkinThickness': df_test['SkinThickness'].median()}, inplace=True)
df_test.fillna({'Insulin': df_test['Insulin'].median()}, inplace=True)
df_test.fillna({'BMI': df_test['BMI'].mean()}, inplace=True)
```

```python
num_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
df_test[num_cols] = scaler.transform(df_test[num_cols])

df_test.head()
```

```python
df_test = pd.DataFrame(scaler.transform(df_test), columns=saved_cols)
```

```python
print("Expected features:", saved_cols)
print("Test features:", list(df_test.columns))
```

```python
X_test = df_test[saved_cols]
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({"Outcome": predictions})
output.to_csv("predictions.csv", index=False)

# from google.colab import files
# files.download("predictions.csv")
```
